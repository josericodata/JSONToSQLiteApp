```python
from faker import Faker
import random
import json
import pandas as pd
```

```python
# Initialize faker
fake = Faker()

# Predefined membership types and statuses
membership_types = ["Basic", "Standard", "Premium"]
subscription_statuses = ["Active", "Inactive", "Suspended"]

# Function to generate a fake gym customer
def generate_gym_customer(id):
    return {
        "id": id,
        "full_name": fake.name(),
        "email": fake.email(),
        "phone_number": fake.phone_number(),
        "age": random.randint(18, 65),
        "gender": random.choice(["Male", "Female", "Other"]),
        "country": fake.country(),
        "join_date": fake.date_between(start_date='-3y', end_date='today').isoformat(),
        "membership_type": random.choice(membership_types),
        "subscription_status": random.choice(subscription_statuses),
        "last_check_in": fake.date_between(start_date='-30d', end_date='today').isoformat(),
        "weight_kg": round(random.uniform(50, 120), 1),
        "height_cm": round(random.uniform(150, 200), 1)
    }

# Generate 100 records
gym_customers = [generate_gym_customer(i+1) for i in range(100)]

# Convert to DataFrame for preview
df_gym = pd.DataFrame(gym_customers)
df_gym.head()
```

```python
# Save as a JSON file
output_file = "gym_customers.json"
with open(output_file, "w") as f:
    json.dump(gym_customers, f, indent=4)

print(f"✅ JSON file saved as: {output_file}")
```
